```python
length = float(0.13)
diameter = float(0.45)
height = float(0.115)
```

```python
if length <= 0.455:
  length_short = 1
  length_medium = 0
  length_long = 0
if length > 0.455 and length < 0.545:
  length_short = (0.545 - length) / (0.545 - 0.455)
  length_medium = (length - 0.455) / (0.545 - 0.455)
  length_long = 0
if length == 0.545:
  length_short = 0
  length_medium = 1
  length_long = 0
if length > 0.545 and length < 0.615:
  length_short = (0.615 - length) / (0.615 - 0.545)
  length_medium = (length - 0.545) / (0.615 - 0.545)
  length_long = 0
if length >= 0.615:
  length_short = 0
  length_medium = 0
  length_long = 1
```

```python
if diameter <= 0.35:
  diameter_small = 1
  diameter_big = 0
if diameter > 0.35 and diameter < 0.48:
  diameter_small = (0.48 - diameter) / (0.48 - 0.35)
  diameter_big = (diameter - 0.35)/(0.48-0.35)
if diameter >= 0.48:
  diameter_small = 0
  diameter_big = 1
```

```python
if height <= 0.115:
  height_short = 1
  height_tall = 0
if height > 0.155 and height < 0.165:
  height_short = (0.165 - height)/(0.165-0.155)
  height_tall = (height-0.155)/(0.165-0.155)
if height <= 0.165:
  height_short = 0
  height_tall = 1
```

```python
print('Maka panjangnya dalam variabel linguistik adalah ')
print('Pendek', length_short)
print('Sedang', length_medium)
print('Panjang', length_long)


print('Maka diameternya dalam variabel linguistik adalah')
print('Kecil', diameter_small)
print('Besar', diameter_big)


print('Maka tinggi dalam variabel linguistik adalah')
print('Pendek', height_short)
print('Tinggi', height_tall)
```

```python
list_predicate_value = []
list_z_value = []


def inference_no_attration(z):
    # Menggunakan metode Tsukamoto
    if z <= 237:
        alpha_min = 0
        z_value = 0
    elif 237 <= z <= 1233:
        alpha_min = (z - 237) / (1233 - 237)
        z_value = 1
    else:
        alpha_min = 1
        z_value = 1

    # Menyimpan nilai pada list
    list_predicate_value.append(alpha_min)
    list_z_value.append(z_value)

    return z_value


def inference_yes_attration(z):
    # Menggunakan metode Tsukamoto
    if z <= 237:
        alpha_min = 1
        z_value = 1
    elif 237 <= z <= 1233:
        alpha_min = (1233 - z) / (1233 - 237)
        z_value = 0
    else:
        alpha_min = 0
        z_value = 0

    # Menyimpan nilai pada list
    list_predicate_value.append(alpha_min)
    list_z_value.append(z_value)

    return z_value
```

```python
def infrant_inference(length, diameter, height):
  pv = min(length, diameter, height)
  list_predicate_value.append(pv)

  if pv == 0:
    list_z_value.append(2)
  if pv == 1:
    list_z_value.append(1)
  if pv > 0 and pv < 1:
    value = 2 - pv
    list_z_value.append(value)
```

```python
def female_inference(length, diameter, height):
  pv = min(length, diameter, height)
  list_predicate_value.append(pv)

  if pv == 0:
    list_z_value.append(2)
  if pv == 1:
    list_z_value.append(3)
  if pv > 0 and pv < 1:
    value = 2 + pv
    list_z_value.append(value)
```

```python
def male_inference(length, diameter, height):
  pv = min(length, diameter, height)
  list_predicate_value.append(pv)

  if pv == 0:
    list_z_value.append(2)
  if pv == 1:
    list_z_value.append(3)
  if pv > 0 and pv < 1:
    value = 2 + pv
    list_z_value.append(value)
```
