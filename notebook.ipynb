## Fuzzifikasi

### 1. Fuzifikasi Umur

```python
def junior(age):
    if age <= 22:
        return 1
    elif 22 < age <= 32:
        return (32 - age) / 10
    elif age > 32:
        return 0

def junior_senior(age):
    if 30 <= age <= 32:
      return (age - 30) / 2
    elif 32 < age < 38:
      return 1
    elif 38 <= age <= 40:
      return (40 - age) / 2
    else:
      return 0
    
def senior(age):
    if age <= 38:
        return 0
    elif 38 < age <= 40:
        return (age - 38) / 2
    elif age > 40:
        return 1

```

### 2. Fuzzifikasi Business Travel

```python
def non_travel(bt):
  if bt < 1:
    return 1
  else:
    return 0
  
  
def rarely_travel(bt):
  if bt == 1:
    return 1
  elif 2 <= bt <= 5:
    return (bt - 2) / 3
  elif bt > 5:
    return 0
  
def frequently_travel(bt):
  if bt < 3:
    return 0
  elif 3 <= bt <= 8:
    return (bt - 3) / 5
  elif bt > 8:
    return 1  


```

### 3. Fuzzifikasi Hourly Rate

```python
def low_rate(hr):
  if hr <= 30:
    return 1
  elif 30 < hr <= 60:
    return (60 - hr) / 30
  elif hr > 60:
    return 0
  
def high_rate(hr):
  if hr < 50:
    return 0
  elif 50 <= hr <= 80:
    return (hr - 50) / 30
  elif hr > 80:
    return 1
  
```

### 3. fuzzifikasi overtime

```python
def no_overtime(ot):
  if ot <= 5:
    return 1
  elif 5 < ot <= 8:
    return (8 - ot) / 3
  elif ot > 8:
    return 0
  
def yes_overtime(ot):
  if ot < 6:
    return 0
  elif 6 <= ot <= 10:
    return (ot - 6) / 4
  elif ot > 10:
    return 1
```

```python
print("Junior: ", junior(input_age))
print("Junior Senior: ", junior_senior(input_age))
print("Senior: ", senior(input_age))

print("Non Travel: ", non_travel(input_business_travel))
print("Rarely Travel: ", rarely_travel(input_business_travel))
print("Frequently Travel: ", frequently_travel(input_business_travel))

print("Low Rate: ", low_rate(input_hourly_rate))
print("High Rate: ", high_rate(input_hourly_rate))

print("No Overtime: ", no_overtime(input_overtime))
print("Yes Overtime: ", yes_overtime(input_overtime))
```

## Inferensi and Rules

```python
def fuzzy_inference(junior_age, junior_senior_age, senior_age,
                    non_travel_bt, rarely_travel_bt, frequently_travel_bt,
                    low_rate_hr, high_rate_hr,
                    no_overtime_ot, yes_overtime_ot):
    # Aturan fuzzy
    rule1 = min(junior_age, non_travel_bt, low_rate_hr, no_overtime_ot)
    rule2 = min(junior_age, rarely_travel_bt, low_rate_hr, no_overtime_ot)
    rule3 = min(junior_age, frequently_travel_bt, low_rate_hr, no_overtime_ot)
    rule4 = min(junior_age, rarely_travel_bt, low_rate_hr, yes_overtime_ot)
    rule5 = min(junior_age, frequently_travel_bt, low_rate_hr, yes_overtime_ot)
    rule6 = min(junior_age, non_travel_bt, high_rate_hr, no_overtime_ot)
    rule7 = min(junior_age, rarely_travel_bt, high_rate_hr, no_overtime_ot)
    rule8 = min(junior_age, frequently_travel_bt, high_rate_hr, no_overtime_ot)
    rule9 = min(junior_age, non_travel_bt, high_rate_hr, yes_overtime_ot)
    rule10 = min(junior_age, rarely_travel_bt, high_rate_hr, yes_overtime_ot)
    rule11 = min(junior_age, frequently_travel_bt, high_rate_hr, yes_overtime_ot)
    rule12 = min(junior_senior_age, non_travel_bt, low_rate_hr, no_overtime_ot)
    rule13 = min(junior_senior_age, rarely_travel_bt, low_rate_hr, no_overtime_ot)
    rule14 = min(junior_senior_age, frequently_travel_bt, low_rate_hr, no_overtime_ot)
    rule15 = min(junior_senior_age, non_travel_bt, low_rate_hr, yes_overtime_ot)
    rule16 = min(junior_senior_age, rarely_travel_bt, low_rate_hr, yes_overtime_ot)
    rule17 = min(junior_senior_age, frequently_travel_bt, low_rate_hr, yes_overtime_ot)
    rule18 = min(junior_senior_age, non_travel_bt, high_rate_hr, no_overtime_ot)
    rule19 = min(junior_senior_age, rarely_travel_bt, high_rate_hr, no_overtime_ot)
    rule20 = min(junior_senior_age, frequently_travel_bt, high_rate_hr, no_overtime_ot)
    rule21 = min(junior_senior_age, non_travel_bt, high_rate_hr, yes_overtime_ot)
    rule22 = min(junior_senior_age, rarely_travel_bt, high_rate_hr, yes_overtime_ot)
    rule23 = min(junior_senior_age, frequently_travel_bt, high_rate_hr, yes_overtime_ot)
    rule24 = min(senior_age, non_travel_bt, low_rate_hr, no_overtime_ot)
    rule25 = min(senior_age, rarely_travel_bt, low_rate_hr, no_overtime_ot)
    rule26 = min(senior_age, frequently_travel_bt, low_rate_hr, no_overtime_ot)
    rule27 = min(senior_age, non_travel_bt, low_rate_hr, yes_overtime_ot)
    rule28 = min(senior_age, rarely_travel_bt, low_rate_hr, yes_overtime_ot)
    rule29 = min(senior_age, frequently_travel_bt, low_rate_hr, yes_overtime_ot)
    rule30 = min(senior_age, non_travel_bt, high_rate_hr, no_overtime_ot)
    rule31 = min(senior_age, rarely_travel_bt, high_rate_hr, no_overtime_ot)
    rule32 = min(senior_age, frequently_travel_bt, high_rate_hr, no_overtime_ot)
    rule33 = min(senior_age, non_travel_bt, high_rate_hr, yes_overtime_ot)
    rule34 = min(senior_age, rarely_travel_bt, high_rate_hr, yes_overtime_ot)
    rule35 = min(senior_age, frequently_travel_bt, high_rate_hr, yes_overtime_ot)

    # Inferensi fuzzy
    result_no = max(rule1, rule2, rule3, rule6, rule7, rule9, rule12, rule13, rule14, rule15, rule16, rule18, rule19, rule20, rule21, rule22, rule23, rule24, rule25, rule26, rule27, rule28, rule29, rule30, rule31, rule32, rule33, rule34, rule35)
    result_yes = max(rule4, rule5, rule8, rule10, rule11, rule17)

    return result_no, result_yes
```

## Defuzzifikasi

```python
def defuzzification(result_no, result_yes):
    centroid_value = (result_yes * 1 + result_no * 0) / (result_yes + result_no)
    return centroid_value
```

## Prediksi Karyawan yang akan keluar

```python
# Prediksi karyawan attrition
def predict_employee_attrition(age, bt, hr, ot):
    # Fuzzifikasi
    junior_value = junior(age)
    junior_senior_value = junior_senior(age)
    senior_value = senior(age)

    non_travel_value = non_travel(bt)
    rarely_travel_value = rarely_travel(bt)
    frequently_travel_value = frequently_travel(bt)

    low_rate_value = low_rate(hr)
    high_rate_value = high_rate(hr)

    no_overtime_value = no_overtime(ot)
    yes_overtime_value = yes_overtime(ot)

    # Inferensi Fuzzy
    result_no, result_yes = fuzzy_inference(junior_value, junior_senior_value, senior_value,
                                            non_travel_value, rarely_travel_value, frequently_travel_value,
                                            low_rate_value, high_rate_value,
                                            no_overtime_value, yes_overtime_value)

    # Defuzzifikasi
    centroid_value = defuzzification(result_no, result_yes)

    # Output
    print('--'*20)
    print("Fuzzifikasi Values:")
    print(f"Junior Value: {junior_value}")
    print(f"Junior Senior Value: {junior_senior_value}")
    print(f"Senior Value: {senior_value}")
    print(f"Non-Travel Value: {non_travel_value}")
    print(f"Rarely Travel Value: {rarely_travel_value}")
    print(f"Frequently Travel Value: {frequently_travel_value}")
    print(f"Low Rate Value: {low_rate_value}")
    print(f"High Rate Value: {high_rate_value}")
    print(f"No Overtime Value: {no_overtime_value}")
    print(f"Yes Overtime Value: {yes_overtime_value}")

    print('--'*20)
    print("\nInferensi Fuzzy Values:")
    print(f"Result No: {result_no}")
    print(f"Result Yes: {result_yes}")


    print('--'*20)
    print("\nDefuzzifikasi Value:")
    print(f"Value: {centroid_value}")

    # Decision
    if centroid_value >= 0.5:
        return "Yes"
    else:
        return "No"
```
