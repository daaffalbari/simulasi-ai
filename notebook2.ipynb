### 1. Himpunan Age(Junior, Junior Senior, Senior)



```python
def age_junior(age):
    if age <= 22 or age >= 32:
        return 0
    elif 22 <= age <= 27:
        return (age - 22) / (27 - 22)
    elif 27 <= age <= 32:
        return (32 - age) / (32 - 27)

def age_juniorsenior(age):
    if age <= 30 or age >= 40:
        return 0
    elif 30 <= age <= 35:
        return (age - 30) / (35 - 30)
    elif 35 <= age <= 40:
        return (40 - age) / (40 - 35)

def age_senior(age):
    if age <= 38 or age >= 60:
        return 0
    elif 38 <= age <= 49:
        return (age - 38) / (49 - 38)
    elif 49 <= age <= 60:
        return (60 - age) / (60 - 49)

# Contoh pengujian\
    
print(f"Junior({age}) =", age_junior(age))

print(f"Junior Senior({age}) =", age_juniorsenior(age))

print(f"Senior({age}) =", age_senior(age))
```

### 2. Himpunan Business Travel(Non-Travel, Travel Rarely, Travel Frequently)


```python
def travel_non(business_travel):
    return 0

def travel_rarely(business_travel):
    if business_travel <= 2 or business_travel >= 5:
        return 0
    elif 2 <= business_travel <= 3.5:
        return (business_travel - 2) / (3.5 - 2)
    elif 3.5 <= business_travel <= 5:
        return (5 - business_travel) / (5 - 3.5)
    
def travel_frequently(business_travel):
    if business_travel <= 3 or business_travel >= 8:
        return 0
    elif 3 <= business_travel <= 5.5:
        return (business_travel - 3) / (5.5 - 3)
    elif 5.5 <= business_travel <= 8:
        return (8 - business_travel) / (8 - 5.5)
    

# Contoh pengujian
print(f"Non({business_travel}) =", travel_non(business_travel))
print(f"Rarely({business_travel}) =", travel_rarely(business_travel))
print(f"Frequently({business_travel}) =", travel_frequently(business_travel))
```

### 3. Himpunan Hourly Rate(Low, High)


```python
def hourlyrate_low(hourlyRate):
  if hourlyRate <= 30 or hourlyRate >= 60:
    return 0
  elif 30 <= hourlyRate <= 45:
    return (hourlyRate - 30) / (45 - 30)
  elif 45 <= hourlyRate <= 60:
    return (60 - hourlyRate) / (60 - 45)
  

def hourlyrate_high(hourlyRate):
  if hourlyRate <= 50 or hourlyRate >= 80:
    return 0
  elif 50 <= hourlyRate <= 65:
    return (hourlyRate - 50) / (65 - 50)
  elif 65 <= hourlyRate <= 80:
    return (80 - hourlyRate) / (80 - 65)
  

# Contoh pengujian
print(f"Low({hourly_rate}) =", hourlyrate_low(hourly_rate))
print(f"High({hourly_rate}) =", hourlyrate_high(hourly_rate))
```

### 4. Himpunan Over Time(No, Yes)


```python
def overtime_no(overtime):
  if overtime <= 5 or overtime >= 8:
    return 0
  elif 5 <= overtime <= 6.5:
    return (overtime - 5) / (6.5 - 5)
  elif 6.5 <= overtime <= 8:
    return (8 - overtime) / (8 - 6.5)
  

def overtime_yes(overtime):
  if overtime <= 6 or overtime >= 10:
    return 0
  elif 6 <= overtime <= 8:
    return (overtime - 6) / (8 - 6)
  elif 8 <= overtime <= 10:
    return (10 - overtime) / (10 - 8)
  

# Contoh pengujian
print(f"No({overtime}) =", overtime_no(overtime))
print(f"Yes({overtime}) =", overtime_yes(overtime))
```

## Kesimpulan

```python
def menentukan_pengurangan_karyawan_yes(z):
    if z <= 237:
        return 0
    elif 237 < z <= 1233:
        return (1233 - z) / (1233 - 237)
    else:
        return 0
    
def menentukan_pengurangan_karyawan_no(z):
    if z <= 237:
        return 1
    elif 237 < z <= 1233:
        return (z - 237) / (1233 - 237)
    else:
        return 1
    

z = nilai_setara
pengurangan_karyawan_yes_val = menentukan_pengurangan_karyawan_yes(z)
pengurangan_karyawan_no_val = menentukan_pengurangan_karyawan_no(z)


print(f"pengurangan_karyawan_yes_val = {pengurangan_karyawan_yes_val}")
print(f"pengurangan_karyawan_no_val = {pengurangan_karyawan_no_val}")

if pengurangan_karyawan_yes_val < pengurangan_karyawan_no_val:
    pengurangan_karyawan = 'Yes'
else:
    pengurangan_karyawan = 'No'

print(f"Hasil = {pengurangan_karyawan}")
```
